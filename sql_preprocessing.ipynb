```python
import pandas as pd
```

```python
df=pd.read_csv('historical_emissions.csv')
df
```

```python
normalized_data = df.melt(
    id_vars=["ISO", "Country", "Data source", "Sector", "Gas", "Unit"],
    var_name="Year",
    value_name="Emissions"
)

# Convert the 'Year' column to integer for easier analysis
normalized_data["Year"] = normalized_data["Year"].astype(int)
```

```python
normalized_data = normalized_data.drop_duplicates()
```

```python
# Remove rows with zero or NaN emissions
normalized_data = normalized_data[(normalized_data["Emissions"] != 0) & (normalized_data["Emissions"].notna())]
```
